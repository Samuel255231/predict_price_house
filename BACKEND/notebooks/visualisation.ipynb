```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
donnee_clean = pd.read_csv("../data/processed/immobilier_nettoyee.csv")
```

### 6. Visualisation 

#### 6.1. Histogrammes (distribution d’une variable numérique)
- visualiser la répartition des valeurs d’une variable numérique
- comprendre sa distribution (symétrique, biaisée, etc.)
- pour détecter rapidement des valeurs aberrantes (outliers)
- analyser la concentration des données

```python
# Distribution de prix en million Ar (histogramme classique)
donnee_clean["prix_million_Ar"].hist(bins=20)
plt.title("Distribution de prix de la maison")
plt.xlabel("Prix en million Ar")
plt.ylabel("Nombre de maison")
plt.show()
```

```python
# Distribution de prix en million Ar (histogramme avec pourcentage)
donnee_clean["prix_million_Ar"].hist(
    bins=20,
    weights = np.ones(len(donnee_clean)) / len(donnee_clean) * 100
)

plt.title("Distribution de prix de la maison")
plt.xlabel("Prix en million Ar")
plt.ylabel("Pourcentage (%)")
plt.show()
```

```python
# Distribution de l'anne de construction (histogramme classique)
donnee_clean["annee_construction"].hist(bins=20)
plt.title("Distribution de l'année de construction")
plt.xlabel("Année de construction")
plt.ylabel("Nombre de maisons")
plt.show()
```

#### 6.2. Bar charts (variables catégorielles)
- visualiser la répartition des catégories
- comparer facilement les catégories
- identifier celles qui ont le plus d’impact ou sont les plus fréquentes dans les données

```python
# Nombre de maisons par localisation
donnee_clean["localisation"].value_counts().plot(kind="bar")
plt.title("Nombre de maisons par localisation")
plt.xlabel("Localisation")
plt.ylabel("Nombre de maisons")
plt.show()
```

```python
# Nombre de maisons par type de connexion
donnee_clean["type_connexion"].value_counts().plot(kind="bar")
plt.title("Nombre de maisons par type de connexion")
plt.xlabel("Type de connexion")
plt.ylabel("Nombre de maisons")
plt.show()
```

#### 6.3. Scatter plots (relation entre 2 variables numériques)
- visualiser la relation entre deux variables numériques
- identifier une tendance ou corrélation (positive, négative ou nulle)
- pour détecter des patterns, regroupements (clusters) ou valeurs aberrantes
- aide à mieux comprendre le comportement des données avant le modèle ML

```python
# Relation entre le prix et l'année de construction (nuage de points)
plt.scatter(donnee_clean["annee_construction"], donnee_clean["prix_million_Ar"])
plt.title("Relation entre le prix et l'année de construction")
plt.xlabel("Année de construction")
plt.ylabel("Prix en million Ar")
plt.show()
```

#### Visualisation avec seaborn

```python
# Histogramme du prix des maisons en million Ar avec Seaborn
# Visualisation de la distribution des prix pour analyser les tendances et détecter les valeurs extrêmes

plt.figure(figsize=(10,6))
sns.histplot(donnee_clean["prix_million_Ar"], bins=20, kde=True)  # kde=False pour ne pas afficher la courbe de densité
plt.title("Distribution de prix de la maison")
plt.xlabel("Prix en million Ar")
plt.ylabel("Nombre de maisons")
plt.show()
```

```python
# Histogramme du prix des maisons en million Ar avec pourcentage
# Visualisation de la distribution des prix en proportion (%) pour comparer facilement les fréquences

plt.figure(figsize=(10,6))
sns.histplot(
    donnee_clean["prix_million_Ar"], 
    bins=20, 
    stat="percent",  # axe Y en pourcentage automatiquement
    kde=True,
)
plt.title("Distribution de prix de la maison (en %)")
plt.xlabel("Prix en million Ar")
plt.ylabel("Pourcentage (%)")
plt.show()
```

```python
# Nuage de points coloré selon la localisation
# Les points sont colorés selon la localisation (urbain, périurbain, rural)
# Cela permet de voir l'impact de la localisation sur le prix

plt.figure(figsize=(10,6))
sns.scatterplot(
    x="annee_construction",
    y="prix_million_Ar",
    hue="localisation",      # couleur selon la localisation
    palette="Set2",          # palette de couleurs agréable
    data=donnee_clean
)
plt.title("Relation entre le prix et l'année de construction par localisation")
plt.xlabel("Année de construction")
plt.ylabel("Prix en million Ar")

plt.legend(title="Localisation")
plt.show()
```

```python
# nuage de points (scatter plot)

plt.scatter(donnee_clean["superficie_m2"], donnee_clean["prix_million_Ar"])
plt.show()
```

```python
# Relation superficie vs prix avec droite de régression

sns.regplot(
    x="superficie_m2",
    y="prix_million_Ar",
    data=donnee_clean
)

plt.title("Relation entre superficie et prix")
plt.xlabel("Superficie (m²)")
plt.ylabel("Prix (million Ar)")
plt.show()
```

```python
# Scatter avec couleur selon nb_chambres

sns.scatterplot(
    x="superficie_m2",
    y="prix_million_Ar",
    hue="nb_chambres",
    data=donnee_clean
)

plt.title("Superficie vs Prix (couleur = nb chambres)")
plt.xlabel("Superficie (m²)")
plt.ylabel("Prix (million Ar)")
plt.show()
```

```python
sns.scatterplot(
    x="superficie_m2",
    y="prix_million_Ar",
    hue="nb_chambres",
    size="nb_etages",
    data=donnee_clean
)

plt.title("Superficie vs Prix (couleur + taille)")
plt.show()
```
