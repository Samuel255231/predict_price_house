NETTOYAGE PAR TYPE

UNIFORMISATION

```python
df['acces_route'] = df['acces_route'].str.strip().str.lower()
df['eau_electricite'] = df['eau_electricite'].str.strip().str.lower()
df['parking'] = df['parking'].str.strip().str.lower()
df['type_sol'] = df['type_sol'].str.strip().str.lower()
df['etat_maison'] = df['etat_maison'].str.strip().str.lower()
df['type_connexion'] = df['type_connexion'].str.strip().str.lower()
df['localisation'] = df['localisation'].str.strip().str.lower()
```

TYPE OBJETS

CREATION MAPPING

```python
mapping = {
    '0': False,
    'no': False,
    'non': False,
    '1': True,
    'yes': True,
    'oui': True
}
```

APPLICATION DE LA TRANSFORMATION

```python
df['acces_route'] = df['acces_route'].map(mapping)
df['eau_electricite'] = df['eau_electricite'].map(mapping)
df['parking'] = df['parking'].map(mapping)
```

VERIFICATION DES VALEURS

VALEURS NEGATIVES

```python
cols = ['prix', 'superficie_m2', 'nb_chambres', 'nb_etages']

for col in cols:
    print(col, (df[col] < 0).sum())
```

MODIFICATIONS DE TYPES

On change les types des colonnes pour avoir des normes et logiques des données

```python
df['localisation'] = df['localisation'].astype('category')
df['type_connexion'] = df['type_connexion'].astype('category')
df['type_sol'] = df['type_sol'].astype('category')
df['etat_maison'] = df['etat_maison'].astype('category')
df['prix'] = df['prix'].astype('int')
df['nb_chambres'] = df['nb_chambres'].astype('int')
df['nb_etages'] = df['nb_etages'].astype('int')
df['annee_construction'] = df['annee_construction'].astype('int')

print(df.dtypes)
```
